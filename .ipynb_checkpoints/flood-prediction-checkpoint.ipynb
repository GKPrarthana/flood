```python
# Split the dataset into features and target variables
X = train_df.drop('FloodProbability', axis=1)
y = train_df['FloodProbability']
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Split the scaled training dataset into training and validation sets using train_test_split:
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=13)
```

```python
print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators =30, random_state=30)
```

```python
LL_rf = rfr.fit(X_train, y_train)
```

```python
print('The training r_sq is: %.2f'% LL_rf.score(X_train, y_train))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

y_pred_val = rfr.predict(X_val)
mse_val = mean_squared_error(y_val, y_pred_val)
mae_val = mean_absolute_error(y_val, y_pred_val)

print('Validation MSE:', mse_val)
print('Validation MAE:', mae_val)
```

```python
# Define the parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [10, 30, 50, 100, 200],
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=30),
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    verbose=2,
    n_jobs=-1
)
```

```python
# Fit GridSearchCV to the training data
grid_search.fit(X_train, y_train)
```

```python
# Print the best parameters and the best score
print('Best Parameters:', grid_search.best_params_)
print('Best R2 Score:', grid_search.best_score_)

# Retrain the model using the best parameters
rfr_best = RandomForestRegressor(**grid_search.best_params_, random_state=30)
rfr_best.fit(X_train, y_train)
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                         huuuuuuuuuuylkoi
# Evaluate the model on the validation set
y_pred_val = rfr_best.predict(X_val)
mse_val = mean_squared_error(y_val, y_pred_val)
mae_val = mean_absolute_error(y_val, y_pred_val)

print('Validation MSE:', mse_val)
print('Validation MAE:', mae_val)

# Scale the test data
X_test_scaled = scaler.transform(test_df)

# Make predictions on the test data
y_pred_test = rfr_best.predict(X_test_scaled)
```
